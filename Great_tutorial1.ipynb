```python
# declair dataframe with 2 features like name and marks
df1 = pd.DataFrame({'Name': ['Anu', 'Kalu', 'AJ'], 'Marks': [100, 200, 300]})
df1
```

```python
df = pd.DataFrame(df1)

# Specify the filename for the CSV file
filename = 'iris.csv'

# Save the DataFrame to a CSV file
df.to_csv(filename, index=False)  # Set index=False to exclude row numbers in the CSV

print(f"Table saved to {filename}")
```

```python
x=np.arange(1,10)
x
```

```python
y=2*x
y
```

```python
# plot the points and draw the line
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.scatter(x,y)
plt.plot(x,y, color='r', linestyle='-.')
plt.grid(True)
plt.show()
```

```python
# Bar plot
student = {'Name': ['Anu', 'Kalu', 'AJ'], 'Marks': [100, 200, 300]}
df = pd.DataFrame(student)
plt.bar(df['Name'], df['Marks'], color='g')
plt.title('Bar Plot')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
# scatter plot
x = [10, 20, 30, 40, 50]
y = [20, 40, 60, 80, 100]
plt.scatter(x, y, color='b', marker='*')
plt.plot(x, y, color='r')
plt.show()
```
